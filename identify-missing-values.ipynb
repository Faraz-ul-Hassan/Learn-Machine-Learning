```python
import pandas as pd # Library for Data manipulation
import matplotlib.pyplot as plt # This library uses for Graphical representation
import seaborn as sns # Library for statistical graphics
```

```python
df = pd.read_csv("cat_breeds_dirty.csv",sep=";") # retrieving the csv file and sep refers to separete because our dataset is contains ';' insted to ','.
```

```python
df.head() # for checking top rows. head(3) the input value shows only 3 rows means equal to input.
```

```python
df.shape   # Gives us the total number of rows and columns
```

```python
df.isnull() # Shows True where value is Null and False where is not null
```

```python
df.isnull().sum() # return count of null values in each column
```

```python
(df.isnull().sum()/df.shape[0])*100 # Percentage of Nullity in each column
```

```python
df.isnull().sum().sum() # Overall Null values
```

```python
(df.isnull().sum().sum()/(df.shape[0]*df.shape[1]))*100 # Overall Null percentage
```
