### --> Ordinal Encoding

## Now Encode our real Dataset

```python
import pandas as pd
```

```python
dataset = pd.read_csv("my_cleaned.csv")
dataset.head(7)
```

```python
dataset.drop(["Unnamed: 0","Allowed_outdoor","Preferred_food"],axis=1,inplace=True)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset[dataset['Fur_colour_dominant'] != 'what does it mean dominant?']
dataset = dataset[dataset['Eye_colour'] != 'I dont know. Its pretty!']
```

```python
dataset['Eye_colour'] = dataset['Eye_colour'].replace('cute', 'unknown')
```

```python
# Clean 'Breed'
dataset['Breed'] = dataset['Breed'].str.lower().replace({
    'ankora': 'angora',
    'angorra': 'angora',
    'my coon': 'maine coon',
    'maine loon': 'maine coon',
    'rack doll': 'ragdoll',
    'wrack doll': 'ragdoll'
})

# Clean 'Country' (merge variations of France)
dataset['Country'] = dataset['Country'].str.lower().replace({
    'la france!!!!': 'france',
    'vive la france!': 'france',
    'my country': 'unknown',  # Handle ambiguous entries
    'where i live': 'unknown',
    'with me': 'unknown'
})
```

```python
cols = dataset.select_dtypes(include='object').columns
cols
```

```python
for i in cols:
    print(dataset[i].unique())
```

```python
# Now we have prettier data
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# 1. Separate column types
cat_cols = dataset.select_dtypes(include='object').columns
num_cols = dataset.select_dtypes(include=['int64', 'float64', 'bool']).columns

# 2. Initialize encoders
ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
le = LabelEncoder()

# 3. Manually binary encode 'Gender' (if exists)
if 'Gender' in cat_cols:
    dataset['Gender'] = dataset['Gender'].map({'female': 0, 'male': 1})
    cat_cols = cat_cols.drop('Gender')

# 4. Choose one-hot encoding columns (low-cardinality)
# You can customize this threshold (e.g. <10 categories)
ohe_cols = [col for col in cat_cols if dataset[col].nunique() <= 10]
label_cols = [col for col in cat_cols if col not in ohe_cols]

# 5. One-hot encode
ohe_encoded = ohe.fit_transform(dataset[ohe_cols])
ohe_df = pd.DataFrame(ohe_encoded, columns=ohe.get_feature_names_out(ohe_cols), index=dataset.index)

# 6. Label encode other categorical columns
for col in label_cols:
    dataset[col + '_encoded'] = le.fit_transform(dataset[col])

# 7. Combine all encoded label cols
label_df = dataset[[col + '_encoded' for col in label_cols]]

# 8. Combine with all numeric columns
final_df = pd.concat([
    dataset[num_cols],
    dataset[['Gender']] if 'Gender' in dataset.columns else pd.DataFrame(),
    ohe_df,
    label_df
], axis=1)

final_df.head()
```

```python
final_df.to_csv('encoded_dataset.csv', index=False)
```
